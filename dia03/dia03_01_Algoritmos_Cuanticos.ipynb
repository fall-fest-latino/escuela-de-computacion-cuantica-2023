# 1. Introduccin

Un **algoritmo cuntico** es un algoritmo que se ejecuta en un modelo realista de computacin cuntica, siendo el modelo ms utilizado el modelo de computacin de [circuitos cunticos](https://en.wikipedia.org/wiki/Quantum_circuit).

Un algoritmo clsico (o no cuntico) es una secuencia finita de instrucciones, o un procedimiento paso a paso para resolver un problema, donde cada paso o instruccin se puede realizar en una computadora clsica. De manera similar, un *algoritmo cuntico* es un procedimiento paso a paso, donde cada uno de los pasos se puede realizar en una computadora cuntica. Aunque todos los algoritmos clsicos tambin se pueden realizar en una computadora cuntica [[Lanzagorta, M. (2009)](https://books.google.com/books?id=-wkJIuw0YRsC&q=quantum%2520computer%2520equivalent%2520classical%2520computer&pg=PA23)], el trmino algoritmo cuntico se usa generalmente para aquellos algoritmos que parecen intrnsecamente cunticos, o usan alguna caracterstica esencial de la computacin cuntica, como la *[superposicin cuntica](https://en.wikipedia.org/wiki/Quantum_superposition)* o el *[entrelazamiento cuntico](https://en.wikipedia.org/wiki/Quantum_entanglement)*.

Lo que hace que los algoritmos cunticos sean interesantes es que podran resolver algunos problemas ms rpido que los algoritmos clsicos porque la superposicin cuntica y el entrelazamiento cuntico que explotan los algoritmos cunticos, probablemente no se pueden simular de manera eficiente en las computadoras clsicas.

## 1.A. Supremaca cuntica

Se le conoce como **[supremaca cuntica](https://en.wikipedia.org/wiki/Quantum_supremacy)** al objetivo de demostrar que un dispositivo cuntico programable puede resolver un problema que ninguna computadora clsica puede resolver en cualquier cantidad de tiempo factible (independientemente de la utilidad del problema).

En Octubre de 2019, Google clam haber alcanzado la **supremaca cuntica** con su computadora cuntica llamada *Sycamore* [[Arute, F., et. al.](https://www.nature.com/articles/s41586-019-1666-5)]; sin embargo, IBM no tard en salir a cuestionar esa afirmacin [[Pednault, E., et. al.](https://arxiv.org/abs/1910.09534)].

Las siguientes son propuestas para demostrar la supremaca cuntica usando la tecnologia actual, usualmente llamada [NISQ](https://en.wikipedia.org/wiki/Noisy_intermediate-scale_quantum_era) (Noisy Intermediate-Scale Quantum):

- un problema computacional bien definido
- un algoritmo cuntico que resuelva este problema
- un algoritmo clsico para comparar el mejor de los casos al resolver el problema
- un argumento terico de la complejidad que, bajo una suposicin razonable, ningn algoritmo clsico puede ser significativamente mejor

Ejemplos de algoritmos a usar en estas propuestas: [Algoritmo de Shor](https://en.wikipedia.org/wiki/Shor%27s_algorithm) (para factorizar enteros), [Muestreo de bosones](https://en.wikipedia.org/wiki/Boson_sampling) (para resolver problemas de muestreo y bsqueda en fotones idnticos), [Muestreo de la distribucin de salida de circuitos cunticos aleatorios](https://en.wikipedia.org/wiki/Cross-entropy_benchmarking) (para simular un circuito cuntico aleatorio arbitrario). Este ltimo fue el usado por Google en su famoso artculo de 2019.

El trmino ha sido ampliamente criticado y actualmente en muchos lados utilizan en su lugar: **Ventaja Cuntica**.

## 1.B. Tiempo de ejecucin

Un objetivo muy general del campo de los algoritmos es resolver una tarea computacional mediante un algoritmo cuyo nmero de pasos (coloquialmente llamado su "tiempo de ejecucin") escala de manera polinomial con el tamao de la entrada, $n$, evitando as la explosin combinatoria. Las tareas computacionales para las que existen tales algoritmos de tiempo polinomial (clsicos) pertenecen a la clase de complejidad $P$. La clase de complejidad correspondiente en la computacin cuntica, tiempo polinomial cuntico de error acotado ([BQP](https://en.wikipedia.org/wiki/BQP), Bounded-error Quantum Polynomial), contiene todas aquellas tareas computacionales que una computadora cuntica sera capaz de resolver en tiempo polinomial. Por el contrario, los algoritmos cuyo tiempo de ejecucin escala exponencialmente (muy rpidamente) con el tamao de la entrada se vuelven prohibitivamente costosos a medida que se escala el tamao de la entrada.


Es importante darse cuenta de que **las computadoras cunticas no aceleran uniformemente todos los problemas computacionales**. Una de las clases ms importantes de problemas computacionales, los problemas [NP-completos](https://en.wikipedia.org/wiki/NP-completeness), se han descrito como la *bsqueda de una aguja en un pajar* pues es exponencialmente grande. Casi al mismo tiempo que el anuncio de Shor (sobre su famoso algoritmo), **[Bennett et al.](https://arxiv.org/abs/quant-ph/9701001) demostr que los algoritmos cunticos requieren un tiempo exponencial para resolver problemas NP-completos en el modelo de caja negra** (es decir, si el algoritmo ignora la estructura detallada del problema) y por lo tanto es poco probable que proporcione aceleraciones exponenciales para tales problemas. Ms precisamente, si $N$ denota el tamao del pajar, [Bennett et al.](https://arxiv.org/abs/quant-ph/9701001) mostr que cualquier algoritmo cuntico para encontrar la aguja debe tomar al menos $N^{1/2}$ pasos. Luego, Grover demostr que existe un algoritmo cuntico que puede encontrar la aguja en $O(N^{1/2})$ pasos. La clase $NP$ se caracteriza por el requisito de que una computadora clsica debera poder verificar la correctitud de una solucin en tiempo polinomial (sin importar cun difcil sea encontrar la solucin). 

Los problemas $NP$-completos son los problemas ms difciles dentro de la clase $NP$ e incluyen el famoso *problema del vendedor viajero*, as como miles de problemas de todos los campos de la ciencia. Es ampliamente conjeturado que $P \neq NP$ (este es uno de los famosos siete [Problemas del Milenio](https://en.wikipedia.org/wiki/Millennium_Prize_Problems)), y que cualquier algoritmo clsico debe requerir $e^n$ pasos para resolver problemas $NP$-completos.

La clase de complejidad [BQP](https://en.wikipedia.org/wiki/BQP) (tiempo polinomial cuntico de error acotado, BQP: Bounded-error Quantum Polynomial) es el conjunto de problemas de decisin que puede resolver una computadora cuntica en tiempo polinomial con una probabilidad de error de como mximo 1/3 para todas las instancias [Nielse, M., Chuang, I.](http://mmrc.amss.cas.cn/tlb/201702/W020170224608149940643.pdf). Es el anlogo cuntico de la clase de complejidad clsica [BPP](https://en.wikipedia.org/wiki/BPP_(complexity)).

Un problema es BQP-completo si est en BQP y cualquier problema en BQP se puede reducir a l en tiempo polinomial. De manera informal, la clase de problemas BQP-completos son aquellos que son tan difciles como los problemas ms difciles de BQP y que una computadora cuntica puede resolver de manera eficiente (con error acotado).

Algunos ejemplos de problemas BQP completos son: [Clculo de invariantes de nudos](https://en.wikipedia.org/wiki/Topological_quantum_field_theory), [Simulacin cuntica](https://en.wikipedia.org/wiki/Hamiltonian_simulation), [Resolver un sistema de ecuaciones lineales](https://en.wikipedia.org/wiki/Quantum_algorithm_for_linear_systems_of_equations).

## 1.C. Algoritmos para computadoras cunticas

El diseo de los **algoritmos cunticos** sigue principios completamente diferentes a los de los algoritmos clsicos. Para empezar, incluso los algoritmos clsicos al pasarlos a su versin cuntica, deben moldearse en una forma especial, como **algoritmos reversibles**, antes de que puedan ejecutarse en una computadora cuntica. Los algoritmos que logran aceleraciones cunticas usan ciertos paradigmas algortmicos cunticos o bloques de construccin que no tienen contrapartes clsicas.

Hay una extensa literatura sobre algoritmos cunticos que se ha desarrollado en ltimo cuarto de siglo desde los primeros algoritmos que han surgido. Todos estos algoritmos se basan en un puado de componentes bsicos cunticos y estn diseados para ejecutarse en una computadora cuntica ideal. Los dispositivos cunticos reales son ruidosos, por lo que se ha desarrollado una teora elaborada de **cdigos de correccin de errores cunticos** y **computacin cuntica tolerante a fallas** para convertir computadoras cunticas ruidosas en computadoras cunticas ideales. Sin embargo, esta conversin genera una sobrecarga tanto en la cantidad de qubits como en el tiempo de ejecucin.

El campo ahora est en la era de los dispositivos cunticos ruidosos de escala intermedia ([NISQ](https://en.wikipedia.org/wiki/Noisy_intermediate-scale_quantum_era) (Noisy Intermediate-Scale Quantum)): la carrera para construir computadoras cunticas que sean lo suficientemente grandes (decenas a cientos o unos pocos miles de qubits) para que no puedan ser simuladas de manera eficiente por una computadora clsica, pero no son tolerantes a fallas y, por lo tanto, no pueden implementar directamente los algoritmos desarrollados para computadoras cunticas ideales. Si bien el enorme inters y la financiacin para construir computadoras NISQ indudablemente ha hecho avanzar el calendario de computadoras cunticas tolerantes a fallas y escalables, an queda mucho trabajo por hacer antes de alcanzar cada hito.

Los **mayores desafos futuros son algortmicos**; a corto plazo, esto incluye la bsqueda de tareas computacionales que las computadoras cunticas puedan acelerar. Desarrollar algoritmos que se ejecuten en computadoras NISQ es tan importante como crear los dispositivos fsicos, ya que sin ambos, la mquina no es til. A ms largo plazo, queda mucho trabajo por hacer en el campo de los algoritmos cunticos para computadoras cunticas ideales (escalables, tolerantes a fallas). Describiremos algunos de los principales bloques de construccin de los algoritmos cunticos, as como los algoritmos conocidos para computadoras cunticas idealizadas que proporcionan aceleraciones sobre los mejores algoritmos clsicos para las mismas tareas computacionales.

El poder de los algoritmos cunticos se deriva en ltima instancia de la complejidad exponencial de los sistemas cunticos: el estado de un sistema de $n$ qubits entrelazados se describe mediante (y, por lo tanto, puede codificar) $N = 2^n$ coeficientes complejos. Adems, la aplicacin de cada compuerta elemental en, digamos, dos qubits actualiza los $2^n$ nmeros complejos que describen al estado, por lo que parece realizar $2^n$ clculos en un solo paso. Por otro lado, al final del clculo, cuando se miden los $n$ qubits, el resultado es solo $n$ bits clsicos. El desafo de disear algoritmos cunticos tiles y ventajosos se deriva de la tensin entre estos dos fenmenos: uno debe **encontrar tareas cuyas soluciones operativas hagan uso de este paralelismo y produzcan un estado cuntico final que tenga una alta probabilidad de devolver informacin valiosa tras la medicin**. Los enfoques exitosos aprovechan el fenmeno de la interferencia cuntica para generar resultados tiles. ### 1.F.b. Circuito cuntico

Debemos tener en dnde usar los qubits y las compuertas cunticas, para eso tenemos al **circuito cuntico**.

<div class="alert alert-block alert-success">
Un circuito cuntico es un modelo de computacin cuntica, similar a los circuitos clsicos, en los que una computacin es una secuencia de compuertas cunticas, mediciones, inicializaciones de qubits a valores conocidos y posiblemente otras acciones.

El conjunto mnimo de acciones que un circuito debe poder realizar en los qubits para permitir la computacin cuntica se conoce como los [criterios de DiVincenzo](https://en.wikipedia.org/wiki/DiVincenzo%27s_criteria).

Este tipo de circuitos se escriben de manera en que el eje horizontal representa al tiempo, comenzando en el lado izquierdo y terminando en el derecho. Las lneas horizontales son qubits, las lneas dobles representan bits clsicos. Los elementos que estn conectados por estas lneas son operaciones realizadas en los qubits, como mediciones o compuertas.

Veamos cmo incluir un qubit en un circuito cuntico:

```python
from qiskit import QuantumRegister

qubit1 = QuantumRegister(1)         # el parmetro indica cuntos qubits queremos

print(qubit1)
```

```python
from qiskit import QuantumCircuit

circuit = QuantumCircuit(qubit1) # crear un circuito cuntico con un qubit

circuit.draw('mpl')              # mostramos la representacin grfica del circuito
```

En el siguiente ejemplo se define un circuito cuntico con 3 qubits, los cuales no tienen que ser creados necesariamente usando la clase `QuantumRegister`, sino que basta con indicar la cantidad de qubits que deseamos tenga el circuito, tambin se definen 2 bits clsicos (lneas dobles), y se aplican 3 compuertas. Veremos que efecto tiene cada una ms adelante.

```python
circ = QuantumCircuit(3, 2) # circuito con 3 qubits y 2 bits clsicos

circ.h(0)           # aplicar compuerta H al qubit 0
circ.cx(0, 1)       # aplicar compuerta CNOT a los qubits 0 y 1
circ.cx(0, 2)       # aplicar compuerta CNOT a los qubits 0 y 2
circ.measure(0, 1)  # medir el qubit 0 en el bit clsico 1
circ.measure(2, 0)  # medir el qubit 2 en el bit clsico 0

circ.draw('mpl')    # mostrar el circuito
```

Como se mencion, el tiempo ocurre de izquierda a derecha, entonces la compuerta $H$ es la que se aplica primero, y despus la compuerta $CNOT_{0,1}$ y por ltimo la compuerta $CNOT_{0,2}$, al final ocurren las mediciones.

Un importante punto a resaltar, es que *todos los qubits comienzan en el estado $\ket{0}$* (parte de los criterios de DiVincenzo).

#### Orden de los qubits

En el circuito anterior se tienen 3 qubits: $q_0$, $q_1$ y $q_2$. Dado el subndice podemos saber cual de ellos es el primero, el segundo y el tercero. Pero al momento de hacer la relacin con las expresiones matemticas, debemos tomar en cuenta que, un estado de tres qubits tendr el siguiente orden (respecto a la convencin de Qiskit),

## 2.A. Sumador completo cuntico

Este algortimo nos servir para ver que podemos implementar algoritmos clsicos en las computadoras cunticas.

Al igual que en la electrnica clsica, donde podemos hacer diferentes tipos de [sumadores binarios](https://en.wikipedia.org/wiki/Adder_(electronics)) como semisumadores, sumadores completos, sumadores con propagacin de acarreo, etctera, tambin se pueden hacer sumadores con circuitos cunticos. En este ejemplo, mostraremos cmo se crea un sumador completo cuntico y cmo acta este sumador en los estados de superposicin.

El *sumador completo* suma a los bits de entrada $A$ y $B$ ms un bit de entrada de acarreo $C_{in}$ y produce los bits de suma y  salida de acarreo $C_{out}$. En la electrnica de control clsica, el sumador completo tiene, por lo tanto, tres entradas y dos salidas.

<p style="font-size:10px;font-style:italic;text-align:right"><a href="https://microcontrollerslab.com/cd4008-4-bit-full-adder-ic-pinout-features-example-datasheet/" target="_blank">Imagen obtenida de MicrocontrollersLab</a></p>

Dado que los circuitos cunticos son reversibles, entonces tienen la misma cantidad de qubits de entrada y salida, por lo que definimos una funcin de $4$ qubits, donde los qubits de entrada son `A`, `B`, `CarryIn` y `[cero]` y los qubits de salida son `A`, `B`, `Sum` y `CarryOut`.

Una posible implementacin de un sumador cuntico completo de 2 bits, usando compuertas $CNOT$ y compuertas *Toffoli* es la siguiente:

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit

A = QuantumRegister(1, 'a')
B = QuantumRegister(1, 'b')
CarryIn = QuantumRegister(1, 'cin-sum')
cero = QuantumRegister(1, 'aux-cout')
adder = QuantumCircuit(A, B, CarryIn, cero)

adder.ccx(0, 1, 3)
adder.cx(0, 1)
adder.ccx(1, 2, 3)
adder.cx(1, 2)
adder.cx(0, 1)

adder.draw('mpl')
```

Ahora probemos el sumador anterior, inicializando los qubits de entrada `a`, `b` y `cin`, con un estado que corresponda a los bits que se desea sumar, con los estados $\ket{0}$ y $\ket{1}$, por ejemplo `a=1`, `b=0` y `cin=1`.

```python
inicializacion = QuantumCircuit(4)

inicializacion.x(0)    # a   = 1
inicializacion.x(2)    # cin = 1
inicializacion.barrier()

inicializacion.draw('mpl')
```

```python
suma = inicializacion.compose(adder)

suma.barrier()

suma.add_register(ClassicalRegister(2, 'res'))

suma.measure([2, 3], [0, 1])

suma.draw('mpl')
```

```python
from qiskit.primitives import Sampler
from qiskit.visualization import plot_histogram

sampler = Sampler()
job_suma = sampler.run(suma)

legend = ["Suma 1 + 0 + 1"]
plot_histogram([job_suma.result().quasi_dists[0]], legend=legend)
```

```python
inicializacion2 = QuantumCircuit(4)
inicializacion2.x(0)    # a   = 1
inicializacion2.x(1)    # b   = 1
inicializacion2.x(2)    # cin = 1
inicializacion2.barrier()

suma2 = inicializacion2.compose(adder)
suma2.barrier()
suma2.add_register(ClassicalRegister(2, 'res'))
suma2.measure([2, 3], [0, 1])
display(suma2.draw('mpl'))

sampler2 = Sampler()
job_suma2 = sampler2.run(suma2, shots=1)  # con una sola ejecucin es suficiente

legend = ["Suma 1 + 1 + 1"]
plot_histogram([job_suma2.result().quasi_dists[0]], legend=legend)
```

Tambin podemos usar estados en superposicin y estados entrelazados, como en el siguiente ejemplo donde se establece `A` y `B` en un **estado de Bell** y `CarryIn` en un estado de superposicin. Esto es similar a establecer la entrada en los estados E1, E2, E7 y E8 al mismo tiempo.

Comparando nuevamente la tabla de verdad con el histograma de amplitudes de probabilidad, vemos que se generan los 4 estados de salida que esperaramos.

```python
inicializacion3 = QuantumCircuit(4)
inicializacion3.h(0)        # a y b en estado de Bell
inicializacion3.cx(0, 1)
inicializacion3.h(2)        # estado en superposin: ( |0> + |1> ) / sqrt(2)
inicializacion3.barrier()

suma3 = inicializacion3.compose(adder)  # el 'adder' es el mismo
suma3.barrier()
suma3.add_register(ClassicalRegister(2, 'res'))
suma3.measure([2, 3], [0, 1])
display(suma3.draw('mpl'))

sampler3 = Sampler()
job_suma3 = sampler3.run(suma3, shots=1000)  # en este caso si necesitamos multiples shots

print(job_suma3.result().quasi_dists)

legend = ["Conteos del resultado de la suma3"]
plot_histogram([job_suma3.result().quasi_dists[0]], legend=legend)
```

Qiskit cuenta con clases que facilitan las sumas de qubits, como las clases: [`CDKMRippleCarryAdder`](https://qiskit.org/documentation/_modules/qiskit/circuit/library/arithmetic/adders/cdkm_ripple_carry_adder.html), [`DraperQFTAdder`](https://qiskit.org/documentation/_modules/qiskit/circuit/library/arithmetic/adders/draper_qft_adder.html), [`VBERippleCarryAdder`](https://qiskit.org/documentation/_modules/qiskit/circuit/library/arithmetic/adders/vbe_ripple_carry_adder.html).

A continuacin tenemos un ejemplo que usa la clase `CDKMRippleCarryAdder` para sumar $3+5$:

```python
from qiskit.circuit.library.arithmetic.adders import CDKMRippleCarryAdder

qiskit_adder = CDKMRippleCarryAdder(3, 'full', 'Sumador completo')

operando1 = QuantumRegister(3, 'o1')  # primer nmero a sumar
operando2 = QuantumRegister(3, 'o2')  # segundo nmero a sumar
aux = QuantumRegister(2, 'a')         # acarreo de entrada y de salida
cr = ClassicalRegister(4)             # para almacenar el resultado en bits clsicos

circ = QuantumCircuit(operando1, operando2, aux, cr)

circ.x([operando1[0], operando1[1]])  # 3 en binario: 011
circ.x([operando2[0], operando2[2]])  # 5 en binario: 101

# El orden esperado por la clase 'CDKMRippleCarryAdder' es: carry_in - 1er operando - 2do operando - carry_out
circ.append(qiskit_adder, [aux[0]] + operando1[0:3] + operando2[0:3] + [aux[1]])

circ.measure(operando2[0:3] + [aux[1]], cr) #  esta clase escribe el resultado en el 2do operando y en el carry_out
display(circ.draw('mpl'))


sampler4 = Sampler()
job_circ = sampler4.run(circ, shots=1000)  # en este caso si necesitamos multiples shots

print(job_circ.result().quasi_dists) # resultado es 8, que en binario es 1000

legend = ["CDKMRippleCarryAdder: 5 + 3"]
plot_histogram([job_circ.result().quasi_dists[0]], legend=legend)
```

## 2.B. Protocolo de teleportacin

Alicia quiere enviar informacin cuntica a Bob. Especficamente, supongamos que quiere enviar un estado cuntico, un qubit: $\ket{\psi}=\alpha\ket{0}+\beta\ket{1}$. Esto implica transmitir informacin sobre $\alpha$ y $\beta$ a Bob.

Existe un teorema en la mecnica cuntica que establece que no es posible simplemente hacer una copia exacta de un estado cuntico desconocido. Esto se conoce como el [teorema de la no clonacin](https://en.wikipedia.org/wiki/No-cloning_theorem). Como resultado de esto, podemos ver que Alicia no puede simplemente generar una copia de $\ket{\psi}$ y drsela a Bob. Solo podemos copiar estados clsicos (no superposiciones).

Sin embargo, al aprovechar dos bits clsicos y un par de qubits entrelazados, Alicia puede transferir su estado $\ket{\psi}$ a Bob. A esto se le conoce como teleportacin porque, al final, Bob tendr a $\ket{\psi}$ y Alicia ya no.

$$$$

Para transferir un estado cuntico, Alicia y Bob deben tener ayuda de un tercero, llammoslo Carlos, que se encargue de enviarles un par de qubits entrelazados. Luego, Alicia realiza algunas operaciones en su qubit, enva los resultados a Bob a travs de un *canal de comunicacin clsico*, y Bob luego realiza algunas operaciones en su extremo para recibir el qubit de Alicia.

Entonces, Alicia tendr dos qubits, en uno estar el estado que desea enviarle a Bob, el otro lo utilizar para mantener un estado entrelazado con el nico qubit de Bob.

Veamos los pasos a seguir:

### Paso 1

Todo comienza con la necesidad de enviar un estado cuntico aleatorio, en particular un qubit, es decir, escrito de la forma:

$$\ket{\psi}=\alpha\ket{0}+\beta\ket{1}$$

Quien lo enva, Alicia, no conoce dicho estado a ser enviado.

Creamos los qubits necesarios para llevar a cabo este protocolo:

```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister

qr_alicia = QuantumRegister(2, name="alicia")  # dos qubits de Alicia
qr_bob = QuantumRegister(1, name="bob")        # un qubit de Bob
crz = ClassicalRegister(1, name="crz")         # 2 bits clasicos
crx = ClassicalRegister(1, name="crx")         # en 2 diferentes registros

circuito_teleportacion = QuantumCircuit(qr_alicia, qr_bob, crz, crx)

circuito_teleportacion.draw('mpl')
```

Su amigo en comn, Carlos, se encarga de preparar un estado entrelazado entre uno de los qubits de Alicia (el qubit 1) y el qubit de Bob. El estado que prepara es un **estado de Bell**. En el lenguaje de circuitos cunticos, la forma de crear un *par de Bell* entre dos qubits es primero transferir uno de ellos a la base $X$ ($\ket{+}$ y $\ket{-}$) usando una compuerta de Hadamard, y luego aplicar una compuerta $CNOT$ en el otro qubit controlado por el que est en la base $X$.

El par entrelazado, usando los qubits $\ket{q_2q_1}=\ket{0_b 0_{a_1}}$ queda como,

$$CNOT_{1,2}\big(I\otimes H\big)\ket{0_b 0_{a_1}}=\frac{1}{\sqrt2}\big(\ket{0_b 0_{a_1}}+\ket{1_b 1_{a_1}}\big)=\ket{e}$$

```python
def crear_estado_bell(circ, a, b):
    circ.h(a)     # poner el qubit a en el estado |+>
    circ.cx(a, b) # CNOT con control el qubit a y objetivo el qubit b
```

```python
crear_estado_bell(circuito_teleportacion, 1, 2) # circuito, qubit a, qubit b

circuito_teleportacion.draw('mpl')
```

Despus de esto, el estado de 3 qubits, incluyendo el estado a ser enviado $\ket{\psi}$, queda como,

$$\begin{aligned}
\ket{e}\otimes\ket{\psi}&=\frac{1}{\sqrt2}\Big(\big(\ket{00}+\ket{11}\big)\otimes \alpha\ket{0} + \big(\ket{00}+\ket{11}\big)\otimes \beta\ket{1}\Big)\\
&=\frac{1}{\sqrt2}\big(\alpha\ket{000}+\alpha\ket{110}+\beta\ket{001}+\beta\ket{111}\big)
\end{aligned}$$

### Paso 2

Ahora, de acuerdo con el protocolo, Alicia aplica la compuerta $CNOT$ en sus dos qubits, usando de control el qubit a ser enviado $\ket{\psi}$, seguido de una compuerta Hadamard en ese mismo qubit. Esto da como resultado el estado:

$$\begin{aligned}
\Big(I\otimes I\otimes H\Big)\Big(I\otimes CNOT_{0,1}\Big)\Big(\ket{e}\otimes\ket{\psi}\Big)&=\Big(I\otimes I\otimes H\Big)\Big(I\otimes CNOT_{0,1}\Big)\frac{1}{\sqrt2}\Big(\alpha\ket{000}+\alpha\ket{110}+\beta\ket{001}+\beta\ket{111}\Big)\\
&=\Big(I\otimes I\otimes H\Big)\frac{1}{\sqrt2}\Big(\alpha\ket{000}+\alpha\ket{110}+\beta\ket{011}+\beta\ket{101}\Big)\\
&=\frac{1}{2}\Big(\alpha\big(\ket{000}+\ket{001}+\ket{110}+\ket{111}\big)+\beta\big(\ket{010}-\ket{011}+\ket{100}-\ket{101}\big)\Big)
\end{aligned}$$

Que puede ser reescrito como:

$$\begin{aligned}
=\frac{1}{2} \Big( & ~~~~~~\big(\alpha\ket{0}+\beta\ket{1}\big)~~\ket{00}\\
&+ ~\big(\alpha\ket{0}-\beta\ket{1}\big)~~\ket{01}\\
&+ ~\big(\alpha\ket{1}+\beta\ket{0}\big)~~\ket{10}\\
&+ ~\big(\alpha\ket{1}-\beta\ket{0}\big)~~\ket{11}~~~~~\Big)\\
\end{aligned}$$

En cdigo queda como sigue:

```python
def compuertas_alicia(circ, psi, a):
    circ.barrier()
    circ.cx(psi, a)
    circ.h(psi)
```

```python
compuertas_alicia(circuito_teleportacion, 0, 1)

circuito_teleportacion.draw('mpl')
```

### Paso 3

Alicia mide los dos primeros qubits (que son de su propiedad) y los enva como dos bits clsicos a Bob. El resultado que Alicia obtiene es siempre uno de los cuatro estados bsicos estndar $\ket{00}$, $\ket{01}$, $\ket{10}$, $\ket{11}$, con igual probabilidad.

Sobre la base de su medicin, el estado de Bob se proyectar a,

$$\ket{00}\rightarrow \big(\alpha\ket{0}+\beta\ket{1}\big)$$
$$\ket{01}\rightarrow \big(\alpha\ket{0}-\beta\ket{1}\big)$$
$$\ket{10}\rightarrow \big(\alpha\ket{1}+\beta\ket{0}\big)$$
$$\ket{11}\rightarrow \big(\alpha\ket{1}-\beta\ket{0}\big)$$

Entonces agregamos las mediciones a los qubits de Alicia:

```python
def mediciones_y_envio(circ, psi, a):
    circ.barrier()
    circ.measure(psi, 0)
    circ.measure(a, 1)
```

```python
mediciones_y_envio(circuito_teleportacion, 0, 1)

circuito_teleportacion.draw('mpl')
```

### Paso 4

Bob, al recibir los bits de Alicia, sabe que puede obtener el estado original aplicando las transformaciones apropiadas en su qubit que una vez fue parte del par de Bell entrelazado.

Las transformaciones que necesita aplicar son:

|                Estado de Bob           | Bits recibidos | Compuerta(s) a aplicar |
|:--------------------------------------:|:------------:|:-------------:|
|$$\big(\alpha\ket{0}+\beta\ket{1}\big)$$|00|$$I$$|
|$$\big(\alpha\ket{0}-\beta\ket{1}\big)$$|01|$$Z$$|
|$$\big(\alpha\ket{1}+\beta\ket{0}\big)$$|10|$$X$$|
|$$\big(\alpha\ket{1}-\beta\ket{0}\big)$$|11|$$ZX$$|

Despus de este paso, Bob habr reconstruido con xito el estado de Alicia.

**Nota**: Esta transferencia de informacin es puramente clsica.

```python
def compuertas_bob(circ, b, crz, crx):
    circ.barrier()
    circ.x(b).c_if(crx, 1)  # Nota: la instruccin c_if solo funciona con simulaciones
    circ.z(b).c_if(crz, 1)  #       para usar condiciones en hardware real (y simuladores), se usa el contexto if_test
```

```python
compuertas_bob(circuito_teleportacion, 2, crz, crx)

circuito_teleportacion.draw('mpl')
```

Al final de este protocolo, el qubit de Alicia ahora se ha teleportado a Bob.

Comprobmoslo ejecutando el protocolo completo y simulndolo:

```python
from qiskit.quantum_info import random_statevector
from qiskit.visualization import plot_histogram, plot_bloch_multivector, array_to_latex

# generamos un estado aleatorio (el que enviar Alicia a Bob y al ser aleatorio ella no lo conoce)
psi = random_statevector(2)

# veamos que de que estado se trata
display(psi.draw('latex', prefix="|\\psi\\rangle ="))

plot_bloch_multivector(psi)
```

```python
from qiskit.extensions import Initialize

# crear una instruccin de inicializacin que coloque al qubit en el estado que generamos aleatoriamente
instruccion_inicializacion = Initialize(psi)
instruccion_inicializacion.label = "init"
```

```python
# El protocolo completo:

# Configuracin inicial
qr_alicia = QuantumRegister(2, name="alicia")  # dos qubits de Alicia
qr_bob = QuantumRegister(1, name="bob")        # un qubit de Bob
crz = ClassicalRegister(1, name="crz")         # 2 bits clasicos
crx = ClassicalRegister(1, name="crx")         # en 2 diferentes registros
circuito_teleportacion = QuantumCircuit(qr_alicia, qr_bob, crz, crx)

# Paso 0
# Iniciar el qubit a enviar
circuito_teleportacion.append(instruccion_inicializacion, [qr_alicia[0]])
circuito_teleportacion.barrier()

# Paso 1
# Crear el par entrelazado
crear_estado_bell(circuito_teleportacion, qr_alicia[1], qr_bob[0]) # circuito, qubit a, qubit b

# Paso 2
# "Enviar" q1 a Alicia y q2 a Bob
compuertas_alicia(circuito_teleportacion, qr_alicia[0], qr_alicia[1])

# Paso 3
# Alicia envia sus bits clsicos a Bob
mediciones_y_envio(circuito_teleportacion, qr_alicia[0], qr_alicia[1])

# Paso 4
# Bob decodifica su qubit
compuertas_bob(circuito_teleportacion, qr_bob[0], crz, crx)

# Finalmente, medir el qubit de Bob (en otro cbit)
circuito_teleportacion.barrier()
bob_med = ClassicalRegister(1, name="bob_med")
circuito_teleportacion.add_register(bob_med)
circuito_teleportacion.measure(qr_bob[0], bob_med)


circuito_teleportacion.draw('mpl')
```

```python
from qiskit_aer.primitives import Sampler   # Nota: este 'Sampler' es distinto al que se haba estado usando
                                            #       ya que permite mediciones intermedias
                                            #       si ejecutando localmente, se debe instalar el paquete qiskit-aer
                                            #       !pip install qiskit-aer
from qiskit.result import marginal_counts

sampler = Sampler()

q = QuantumCircuit(1)                                   # circuito temporal para simular solo el estado inicial
q.append(instruccion_inicializacion, [0])
q.measure_all()
job_inicializacion = sampler.run(q)

job_teleportacion = sampler.run(circuito_teleportacion) # simulamos el circuito de teleportacin


# marginar conteos, es decir, solo quedarnos con los del cbit que contiene la medicin de Bob
conteos = job_teleportacion.result().quasi_dists[0].binary_probabilities()
tele_counts = marginal_counts(conteos, [2])


# graficamos los resultados con un histograma

legend = ["Estado original", "Estado teleportado"]
plot_histogram([job_inicializacion.result().quasi_dists[0].binary_probabilities(), tele_counts], legend=legend)
```

## 2.C. Protocolo de codificacin superdensa

### Diferencia entre la Codificacin superdensa y la Teleportacin cuntica

La teleportacin cuntica y la codificacin superdensa estn estrechamente relacionadas, para evitar confusiones, debemos aclarar la diferencia.

La teleportacin cuntica es un proceso mediante el cual el estado de un qubit ($\ket{\psi}$) se puede transmitir de un lugar a otro, utilizando dos bits de comunicacin clsica y un par de Bell. En otras palabras, podemos decir que es un protocolo que destruye el estado cuntico de un qubit en una ubicacin y lo recrea en un qubit en una ubicacin distante, con la ayuda del entrelazamiento. La **codificacin superdensa** es un procedimiento que le permite a alguien enviar dos bits clsicos a otra parte utilizando solo un qubit de comunicacin.

|           Teleportacin       |     Codificacin superdensa     |
|:--------------------------------------:|:------------:|
| Transmite un qubit <br> usando dos bits <br> clsicos | Transmite dos bits <br> clsicos usando <br> un qubit |


El protocolo de teleportacin se puede considerar como una versin invertida del protocolo de codificacin superdensa, en el sentido de que Alicia y Bob simplemente "intercambian su equipo".

### Paso 1

El proceso comienza con una tercera persona, a quien llamaremos Carlos, quien prepara dos qubits en un estado entrelazado. Inicialmente inicia los 2 qubits en el estado base $\ket{0}$. Aplica aplica la compuerta de Hadamard ($H$) al primer qubit para crear una superposicin. Luego aplica la compuerta $CNOT$ ($CX$) usando el primer qubit como control y el segundo como objetivo. Este es el estado entrelazado (par de Bell) que mencionamos anteriormente.

Comenzamos con el estado:

$$\ket{00}=\ket{0}_A\otimes\ket{0}_B$$

donde el qubit que se enviar a Alicia est etiquetado con $A$ y el qubit que se enviar a Bob est etiquetado como $B$. Carlos primero aplica una compuerta de Hadamard al primer qubit, lo que crea una superposicin y obtenemos el estado:

$$\ket{+0}=\frac{1}{\sqrt2}\Big(\ket{00}+\ket{10}\Big)$$

Entonces Carlos aplica la compuerta $CNOT$, esta entrelaza ambos qubits, es decir, invierte el qubit objetivo si el control es $\ket{1}$. Ten en cuenta que el qubit de control es nuestro qubit ms a la izquierda.

$$CNOT\frac{1}{\sqrt2}\Big(\ket{00}+\ket{10}\Big)=\frac{1}{\sqrt2}\Big(\ket{00}+\ket{11}\Big)$$

```python
def crear_par_de_bell():
    qc = QuantumCircuit(2)
    qc.h(1)
    qc.cx(1, 0)
    return qc
```

### Paso 2

Carlos enva el primer qubit a Alicia y el segundo a Bob. El objetivo del protocolo es que Alicia enve 2 bits clsicos de informacin a Bob usando su qubit. Pero antes de hacerlo, necesita aplicar un conjunto de compuertas cunticas a su qubit dependiendo de los 2 bits de informacin que quiera enviar:

| Mensaje a enviar | Compuerta a aplicar | Estado resultante $\Big(\cdot\frac{1}{\sqrt2}\Big)$ |
|:--------------------------------------:|:------------:|:-------------:|
|00|$$I$$|$$\ket{00}+\ket{11}$$
|01|$$X$$|$$\ket{10}+\ket{01}$$
|10|$$Z$$|$$\ket{00}-\ket{11}$$
|11|$$ZX$$|$$-\ket{10}+\ket{01}$$

Por lo tanto, si Alicia quiere enviar un $00$, no le hace nada al qubit (aplica la compuerta de identidad ($I$)). Si quiere enviar un $01$, entonces aplica la compuerta $X$. Dependiendo de lo que quiera enviar, aplica la compuerta adecuada y luego enva su qubit a Bob para el paso final del proceso.

```python
def codificar_mensaje(qc, qubit, msg):
    if len(msg) != 2 or not set(msg).issubset({"0", "1"}): # el mensaje debe ser de solo 0s y 1s
        raise ValueError(f"El mensaje '{msg}' es invlido")
    if msg[1] == "1":
        qc.x(qubit)
    if msg[0] == "1":
        qc.z(qubit)
    return qc
```

### Paso 3

Bob recibe el qubit de Alicia (el qubit ms a la izquierda) y usa su qubit para decodificar el mensaje de Alicia. Ten en cuenta que no necesita tener conocimiento del estado para decodificarlo, simplemente usa la operacin de restauracin.

Bob aplica una compuerta CNOT usando el qubit ms a la izquierda como control y el ms a la derecha como objetivo. Luego aplica una compuerta Hadamard y finalmente realiza una medicin en ambos qubits para extraer el mensaje de Alicia.

| Bob recibe $\Big(\cdot\frac{1}{\sqrt2}\Big)$ | Despus de la compuerta $CNOT$ | Despus de la compuerta $H$ |
|:--------------------------------------:|:------------:|:-------------:|
|$$\ket{00}+\ket{11}$$|$$\ket{00}+\ket{10}$$|$$\ket{00}$$
|$$\ket{10}+\ket{01}$$|$$\ket{11}+\ket{01}$$|$$\ket{01}$$
|$$\ket{00}-\ket{11}$$|$$\ket{00}-\ket{10}$$|$$\ket{10}$$
|$$-\ket{10}+\ket{01}$$|$$-\ket{11}+\ket{01}$$|$$\ket{11}$$

```python
def decodificar_mensaje(qc):
    qc.cx(1, 0)
    qc.h(1)
    return qc
```

```python
# El protcolo completo

# Carlos crea un par entrelazado entre Alicia y Bob
qc = crear_par_de_bell()

# Agregamos una barrera para una separacin visual
qc.barrier()

# En este punto, el qubit 0 va a Alicia y el qubit 1 va a Bob

# Luego, Alicia codifica su mensaje en el qubit 1. En este caso, 
# queremos enviar el mensaje '10'. Puedes intentar cambiar este
# valor y ver cmo afecta al circuito
mensaje = '10'
qc = codificar_mensaje(qc, 1, mensaje)
qc.barrier()
# Luego Alicia enva su qubit a Bob

# Despus de recibir el qubit 0, Bob aplica el protocolo de recuperacin:
qc = decodificar_mensaje(qc)

# Finalmente, Bob mide sus qubits para leer el mensaje de Alicia
qc.measure_all()

# Mostramos el circuiti
qc.draw('mpl')
```

```python
from qiskit.primitives import Sampler

sampler = Sampler()
job_qc = sampler.run(qc, shots=1000)  # usamos multiples shots

print(job_qc.result().quasi_dists)

legend = ["Mensaje enviado"]
plot_histogram([job_qc.result().quasi_dists[0].binary_probabilities()], legend=legend)
```

## 2.D. Algoritmo de Deutsch  Jozsa

El **algoritmo de Deutsch-Jozsa** es un algoritmo cuntico determinista propuesto por David Deutsch y Richard Jozsa en 1992 [[Deutsch, D., Jozsa, R.](https://royalsocietypublishing.org/doi/10.1098/rspa.1992.0167)]. Aunque actualmente tiene poco uso prctico, es uno de los primeros ejemplos de un algoritmo cuntico que es exponencialmente ms rpido que cualquier algoritmo clsico determinista posible.

<a id="2.4.1"></a>
### 2.D.a. El problema

Se tiene una funcin booleana oculta $f$, conocida como *caja negra* o como **orculo**, que toma como entrada una cadena de bits y devuelve $0$ o $1$, es decir:

$$f\big(\{x_0, x_1, x_2, \cdots\}\big)\rightarrow0 \text{ o } 1 \text{, donde } x_n \text{ es } 0 \text{ o } 1$$

La propiedad de la funcin booleana dada es que se garantiza que sea **balanceada** o **constante**. Una funcin *constante* devuelve siempre $0$ o siempre $1$ para cualquier entrada, mientras que una funcin *balanceada* devuelve $0$ para exactamente la mitad de todas las entradas y $1$ para la otra mitad. Nuestra tarea es determinar si la funcin dada es balanceada o constante.

Se debe tener en cuenta que el problema de Deutsch-Jozsa es una extensin de $n$ bits del problema de Deutsch de un solo bit [[Deutsch, D.](https://royalsocietypublishing.org/doi/10.1098/rspa.1985.0070)].

<a id="2.4.2"></a>
### 2.D.b. Solucin clsica

En el mejor de los casos, en una solucin clsica, se necesitan dos consultas al *orculo* para determinar si la funcin booleana oculta $f(x)$ es balanceada, por ejemplo si las siguientes consultas regresan el valor dado:

$$f(0, 0, 0, 0, \cdots)\rightarrow 0$$
$$f(1, 0, 0, 0, \cdots)\rightarrow 1$$

entonces sabremos que la funcin es *balanceada* pues estamos obteniendo dos resultados distintos.

En el peor de los casos, si al hacer consultas para diferentes entradas obtenemos el mismo resultado, tendramos que revisar exactamente la mitad de las posibles entradas ms una, para poder tener la certeza de que $f(x)$ es constante.

Dado que el total de posibles entradas es $2^n$, se tiene que es necesario probar con $2^{n-1}+1$ entradas para tener la certeza de que el orculo es constante, en el peor de los casos.

<div class="alert alert-block alert-info">
<strong>Ejemplo</strong>: 
    
Para cadenas de $4$ bits, tenemos que $n=4$, entonces la cantidad de entradas ser $2^4=16$.
    
Por lo que si revisamos $8$ combinaciones de las $16$ posibles y obtenemos siempre la salida igual a $0$, an es posible que la $9^a$ combinacin regrese un $1$, significando que $f(x)$ es *balanceada*.

$$f(0,0,0,0)\rightarrow 0$$
$$f(0,0,0,1)\rightarrow 0$$
$$f(0,0,1,0)\rightarrow 0$$
$$f(0,0,1,1)\rightarrow 0$$
$$f(0,1,0,0)\rightarrow 0$$
$$f(0,1,0,1)\rightarrow 0$$
$$f(0,1,1,0)\rightarrow 0$$
$$f(0,1,1,1)\rightarrow 0$$
$$$$
$$f(1,0,0,0)\rightarrow 1$$

Probabilsticamente hablando, este es un evento muy improbable. De hecho, si obtenemos el mismo resultado continuamente en sucesin, podemos expresar la probabilidad de que la funcin sea constante en funcin de $k$ entradas como,

$$P_{\text{constante}}(k)=1-\frac{1}{2^{k-1}} ~~~~~~~~~~~~~~ \text{para } 1<k\leq 2^{n-1}$$

Siendo realistas, podramos optar por truncar nuestro algoritmo clsico de manera temprana (sin revisar todas las posibles entradas), digamos si tuviramos ms de un $x\%$ de confianza. Pero si queremos estar $100\%$ seguros, necesitaramos verificar $2^{n-1}+1$ entradas.

<a id="2.4.3"></a>
### 2.D.c. Solucin cuntica

Usando una computadora cuntica, podemos resolver este problema con un $100\%$ de confianza despus de **una sola llamada a la funcin $f(x)$**, siempre que tengamos la funcin $f$ implementada como un **orculo cuntico**, que asigna el estado $\ket{x}\ket{y}$ al estado $\ket{x}\ket{y\oplus f(x)}$, donde $\oplus$ es la [suma mdulo 2](https://en.wikipedia.org/wiki/Modular_arithmetic) (que tambin se puede ver como una compuerta $XOR$ cuntica implementada como una compuerta $CNOT$).

A continuacin se muestra el circuito genrico para el algoritmo de Deutsch-Jozsa.

Veamos los pasos y por cada uno revisemos su implementacin en Qiskit, el ejemplo en cdigo ser para $n=3$, por lo que tendremos un total de $n+1=4$ qubits.

Antes de ver cada paso, generemos dos orculos, uno constante y otro balanceado, para comprobar que el algoritmo funciona correctamente.

```python
n = 3           # cantidad de bits para la funcin f
```

```python
from qiskit import QuantumCircuit
import numpy as np


#---------------------
## Orculo Constante
#---------------------

const_oraculo = QuantumCircuit(n+1)   # +1 qubit que corresponde al segundo registro

# en este caso, la entrada no tiene efecto en la salida, por lo que simplemente 
# configuramos aleatoriamente el qubit de salida en 0 o 1:

output = np.random.randint(2)
if output == 1:
    const_oraculo.x(n)          # aplicar X al qubit de salida (un solo resultado para todas las entradas)

const_oraculo.draw('mpl')
```

```python
#---------------------
## Orculo Balanceado
#---------------------

balan_oraculo = QuantumCircuit(n+1)


# podemos crear un orculo balanceado usando CNOTs con cada qubit de entrada como control
# y el bit de salida como objetivo. Podemos variar los estados de entrada que dan 0 o 1 
# encerrando algunos de los controles con compuertas X. Primero elijamos una cadena binaria de 
# longitud n que dicte qu controles ajustar:

b_str = "101"

# agregar las compuertas X en la poscin donde hay un 1
for qubit in range(len(b_str)):
    if b_str[qubit] == '1':
        balan_oraculo.x(qubit)
balan_oraculo.barrier()

# ahora, aplicamos las compuertas CNOT, usando cada qubit de entrada como control y el 
# qubit de salida como objetivo:
for qubit in range(n):
    balan_oraculo.cx(qubit, n)

balan_oraculo.barrier()

# finalmente, aplicamos compuertas X para terminar de envolver los controles:
for qubit in range(len(b_str)):
    if b_str[qubit] == '1':
        balan_oraculo.x(qubit)

balan_oraculo.draw('mpl')
```

#### Paso 1

Se preparan dos registros cunticos. El primero en un registro de $n$ qubits inicializado en $\ket{0}^{\otimes n}$, el segundo registro es un qubit inicializado en $\ket{1}$,

$$\ket{\psi_0} = \ket{0}^{\otimes n} \, \ket{1}$$

```python
dj_circuit = QuantumCircuit(n+1, n)          # n+1=4 qubits, n=3 bits clsicos

dj_circuit.x(n)                              # colocar el ltimo qubit (segundo registro) en el estado |1>

dj_circuit.barrier()
dj_circuit.draw('mpl')
```

#### Paso 2

Se aplican compuertas Hadamard a cada qubit,

$$\ket{\psi_1}=\frac{1}{\sqrt{2^{n+1}}}\sum_{x=0}^{2^n-1} \, \ket{x} \, \big(\ket{0}-\ket{1}\big)$$

```python
# Aplicar las compuertas H al primer registro
for qubit in range(n):
    dj_circuit.h(qubit)

# Aplicar H al segundo registro, H|1> = |->
dj_circuit.h(n)

dj_circuit.barrier()
dj_circuit.draw('mpl')
```

#### Paso 3

Aplicar el orculo al estado $\ket{x}\ket{y}$ obteniendo el estado $\ket{x}\ket{y\oplus f(x)}$, es decir,

$$
\begin{aligned}
\ket{\psi_2}&=\frac{1}{\sqrt{2^{n+1}}}\sum_{x=0}^{2^n-1} \, \ket{x} \, \big(0\oplus \ket{f(x)}-\ket{1\oplus f(x)}\big)\\
&=\frac{1}{\sqrt{2^{n+1}}}\sum_{x=0}^{2^n-1} \, \ket{x} \, \big(\ket{f(x)}-\ket{1\oplus f(x)}\big)\\
&=\frac{1}{\sqrt{2^{n+1}}}\sum_{x=0}^{2^n-1} \, \big(-1\big)^{f(x)} \, \ket{x} \, \big(\ket{0}-\ket{1}\big)\\
\end{aligned}
$$

ya que para cada $x$, la funcin $f(x)$ es un $0$ o un $1$.

En el cdigo, creamos una funcin que nos ayuda a agregar el orculo deseado de acuerdo a un parmetro:

```python
def dj_oraculo(dj_circuit, caso):
    
    if caso == 'balanceado':
        circuit_return = dj_circuit.compose(balan_oraculo)
    else:
        circuit_return = dj_circuit.compose(const_oraculo)
        
    circuit_return.barrier()
    
    return circuit_return
```

```python
# generemos un circuito con el orculo constante
dj_circ_constante = dj_oraculo(dj_circuit, 'constante')

# generemos un circuito con el orculo balanceado
dj_circ_balanceado = dj_oraculo(dj_circuit, 'balanceado')


dj_circ_balanceado.draw('mpl')
```

#### Paso 4

En este punto, el segundo registro (de un qubit en el estado $\ket{-}$) puede ser ignorado, pues solo interesan las mediciones del primer registro, el estado en el que nos fijaremos es el siguiente,

$$\frac{1}{\sqrt{2^n}}\sum_{x=0}^{2^n-1} \, \big(-1\big)^{f(x)} \, \ket{x}$$

Se aplican compuertas Hadamard a cada qubit del primer registro,

$$
\begin{aligned}
\ket{\psi_3}&=\frac{1}{2^{n}}\sum_{x=0}^{2^n-1} \, \big(-1\big)^{f(x)} \, \left[\sum_{j=0}^{2^n-1} \, \big(-1\big)^{x\cdot j} \, \ket{j} \right]\\
&=\frac{1}{2^{n}}\sum_{j=0}^{2^n-1} \, \left[\sum_{x=0}^{2^n-1} \, \big(-1\big)^{f(x)} \big(-1\big)^{x\cdot j}\right] \, \ket{j}\\
\end{aligned}
$$

donde $x\cdot j=x_0j_0\oplus x_1j_1\oplus \cdots\oplus x_{n-1}j_{n-1}$ es la suma del producto bit a bit.

```python
#---------------------
## Orculo Balanceado
#---------------------

# Aplicar nuevamente compuertas H al primer registro (en el circuito que ya tiene el orculo)
for qubit in range(n):
    dj_circ_balanceado.h(qubit)
    
dj_circ_balanceado.barrier()
dj_circ_balanceado.draw('mpl')
```

```python
#---------------------
## Orculo Constante
#---------------------

# Aplicar nuevamente compuertas H al primer registro (en el circuito que ya tienen el orculo)
for qubit in range(n):
    dj_circ_constante.h(qubit)

dj_circ_constante.barrier()
dj_circ_constante.draw('mpl')
```

#### Paso 5

Medir el primer registro.

Notemos que la probabilidad para que un estado $\ket{k}$ sea medido, est dada por,

$$\left|\frac{1}{2^{n}}\sum_{x=0}^{2^n-1} \, \big(-1\big)^{f(x)} \big(-1\big)^{x\cdot k}\right|^2$$

En particular para el caso $k=0$, o escrito en forma de estado cuntico: $\ket{k}=\ket{0}^{\otimes n}$, la probabilidad sera,

$$\left|\frac{1}{2^{n}}\sum_{x=0}^{2^n-1} \, \big(-1\big)^{f(x)} \right|^2$$

que ser igual a $1$ si $f(x)$ es **constante** (habr interferencia constructiva), e igual a $0$ si $f(x)$ es balanceada (habr interferencia destructiva).

$$$$

En otras palabras, la medicin final ser $\ket{0}^{\otimes n}$ (es decir, todos ceros) si $f(x)$ es constante, y se obtendr cualquier otro estado si $f(x)$ es balanceada.

```python
#---------------------
## Orculo Balanceado
#---------------------

# agregamos mediciones al primer registro
for i in range(n):
    dj_circ_balanceado.measure(i, i)
    
dj_circ_balanceado.draw('mpl')
```

```python
#---------------------
## Orculo Constante
#---------------------

# agregamos mediciones al primer registro
for i in range(n):
    dj_circ_constante.measure(i, i)
    
dj_circ_constante.draw('mpl')
```

**Nota**: El orden de los qubits en el cdigo en Qiskit est invertido respecto al orden en la expresiones matemticas, pero esto no repreenta un problema en los resultados, todo funciona como es esperado, solo se debe tener cuidado al mapear lo definido matemticamente con lo implementado en cdigo.

#### Simulacin

```python
from qiskit.primitives import Sampler
from qiskit.visualization import plot_histogram

#---------------------
## Orculo Constante
#---------------------

sampler = Sampler()
job_const = sampler.run(dj_circ_constante)

legend = ["Funcin constante"]
plot_histogram([job_const.result().quasi_dists[0].binary_probabilities()], legend=legend)
```

```python
#---------------------
## Orculo Balanceado
#---------------------

sampler = Sampler()
job_balan = sampler.run(dj_circ_balanceado)

legend = ["Funcin balanceada"]
plot_histogram([job_balan.result().quasi_dists[0].binary_probabilities()], legend=legend)
```

### 2.D.d. Por qu funciona?

#### Orculo Constante

Cuando el orculo es constante, no tiene efecto (hasta una fase global) en los qubits de entrada, y los estados cunticos antes y despus de consultar el orculo son los mismos. Dado que la compuerta $H$ es su propio inverso, en el *paso 4* invertimos el *paso 2* para obtener el estado cuntico inicial $\ket{00\cdots 0}$ en el primer registro.

$$H^{\otimes n}\mymatrix{r}{ 1 \\ 0 \\ 0 \\ \vdots \\ 0 }=\frac{1}{\sqrt{2^n}}\mymatrix{r}{ 1 \\ 1 \\ 1 \\ \vdots \\ 1 }\xrightarrow{\text{despus de } U_f}H^{\otimes n}\frac{1}{\sqrt{2^n}}\mymatrix{r}{ 1 \\ 1 \\ 1 \\ \vdots \\ 1 }=\mymatrix{r}{ 1 \\ 0 \\ 0 \\ \vdots \\ 0 }$$

$$$$
$$$$

#### Orculo Balanceado

Despus del *paso 2*, el registro de entrada es una superposicin igual de todos los estados en la base computacional. Cuando el orculo es balanceado, el [retroceso de fase (phase kickback)](https://qiskit.org/textbook/ch-gates/phase-kickback.html) agrega una fase negativa a exactamente la mitad de estos estados:

$$U_f\frac{1}{\sqrt{2^n}}\mymatrix{r}{ 1 \\ 1 \\ 1 \\ \vdots \\ 1 }=\frac{1}{\sqrt{2^n}}\mymatrix{r}{ -1 \\ 1 \\ -1 \\ \vdots \\ 1 }$$

El estado cuntico despus de consultar el orculo es ortogonal al estado cuntico antes de consultar el orculo. Por lo tanto, en el *paso 4*, al aplicar las compuertas H, debemos terminar con un estado cuntico que sea ortogonal a $\ket{00\cdots 0}$. Esto significa que nunca debemos medir el estado cero.

## 2.E. Algoritmo de Bernstein  Vazirani

El **algoritmo de BernsteinVazirani**, introducido por primera vez por Ethan Bernstein y Umesh Vazirani en 1997 [E. Bernstein, U. Vazirani](https://epubs.siam.org/doi/10.1137/S0097539796300921), puede verse como una extensin del algoritmo de Deutsch-Jozsa que cubrimos en la seccin anterior. Demuestra que puede haber ventajas en el uso de una computadora cuntica como herramienta computacional para problemas ms complejos que el problema de Deutsch-Jozsa.

<a id="2.5.1"></a>
### 2.E.a. El problema

Nuevamente se proporciona una funcin de caja negra $f$, que toma como entrada una cadena de bits ($x$) y devuelve $0$ o $1$, es decir:

$$f(\{x_0,x_1,x_2,...\}) \rightarrow 0 \textrm{ o } 1 \textrm{ donde } x_n \textrm{ es }0 \textrm{ o } 1 $$

En lugar de que la funcin sea balanceada o constante como en el problema de Deutsch-Jozsa, ahora se garantiza que la funcin devolver el producto bit a bit de la entrada con alguna cadena, $s$. En otras palabras, dada una entrada $x$, $f(x) = s \cdot x \, \text{(mod 2)}$. Se espera que encontremos $s$. Como circuito reversible clsico, el orculo de Bernstein-Vazirani se ve as:

<a id="2.5.2"></a>
### 2.E.b. Solucin clsica

Clsicamente, el orculo devuelve:

$$f_s(x) = s \cdot x \mod 2$$

dada una entrada $x$. Por lo tanto, la cadena de bits oculta $s$ puede revelarse consultando al orculo con la secuencia de entradas:

|Entrada(x)|
|:-----:|
|100...0|
|010...0|
|001...0|
|000...1|

Donde cada consulta revela un bit diferente de $s$ (el bit $s_i$). Por ejemplo, con `x = 1000...0` se puede obtener el bit menos significativo de $s$, con `x = 0100...0` podemos encontrar el siguiente bit menos significativo, y as sucesivamente. Esto significa que necesitaramos llamar a la funcin $f_s(x)$, $n$ veces.

<a id="2.5.3"></a>
### 2.F.c. Solucin cuntica

Usando una computadora cuntica, podemos resolver este problema con un 100% de confianza despus de una sola llamada a la funcin $f(x)$. El algoritmo cuntico de Bernstein-Vazirani para encontrar la cadena de bits oculta es muy simple:

1. Inicializar los qubits de entrada con el estado $|0\rangle^{\otimes n}$ y el qubit de salida con $|{-}\rangle$
2. Aplicar compuertas Hadamard al registro de entrada
3. Consultar el orculo
4. Aplicar compuertas Hadamard al registro de entrada
5. Medir

Para explicar el algoritmo, veamos ms de cerca lo que sucede cuando aplicamos una compuerta H a cada qubit. Si tenemos un estado de $n$-qubits, $|a\rangle$, y aplicamos las compuertas H, veremos la transformacin:

$$
|a\rangle \xrightarrow{H^{\otimes n}} \frac{1}{\sqrt{2^n}} \sum_{x\in \{0,1\}^n} (-1)^ {a\cdot x}|x\rangle.
$$

### Explicacin de la ecuacin

Recordamos que la compuerta Hadamard realiza las siguientes transformaciones en un qubit:

$$ H|0\rangle = \tfrac{1}{\sqrt{2}}(|0\rangle + |1\rangle) $$
$$ H|1\rangle = \tfrac{1}{\sqrt{2}}(|0\rangle - |1\rangle) $$

Usando la notacin de suma, podramos reescribirlo as:

$$ H|a\rangle = \frac{1}{\sqrt{2}}\sum_{x\in \{0,1\}} (-1)^{a\cdot x}|x\rangle. $$

En el caso de dos qubits, se aplica una Hadamard a cada uno, lo que realiza las siguientes transformaciones:

$$ H^{\otimes 2}|00\rangle = \tfrac{1}{2}(|00\rangle + |01\rangle + |10\rangle + |11\rangle) $$
$$ H^{\otimes 2}|01\rangle = \tfrac{1}{2}(|00\rangle - |01\rangle + |10\rangle - |11\rangle) $$
$$ H^{\otimes 2}|10\rangle = \tfrac{1}{2}(|00\rangle + |01\rangle - |10\rangle - |11\rangle) $$
$$ H^{\otimes 2}|11\rangle = \tfrac{1}{2}(|00\rangle - |01\rangle - |10\rangle + |11\rangle) $$

Podemos expresar esto usando la siguiente suma:

$$ H^{\otimes 2}|a\rangle = \frac{1}{2}\sum_{x\in \{0,1\}^2} (-1)^{a\cdot x}| x\rangle $$

<br><br>

En particular, cuando comenzamos con un registro cuntico $|00\dots 0\rangle$ y le aplicamos $n$ compuertas Hadamard, tenemos la siguiente superposicin cuntica:

$$
|00\dots 0\rangle \xrightarrow{H^{\otimes n}} \frac{1}{\sqrt{2^n}} \sum_{x\in \{0,1\}^n} |x \rangle
$$

En este caso, el trmino de fase $(-1)^{a\cdot x}$ desaparece, ya que $a=0$, y por lo tanto $(-1)^{a\cdot x} = 1$.

El orculo clsico $f_s$ devuelve $1$ para cualquier entrada $x$ tal que $s \cdot x\mod 2 = 1$, y devuelve $0$ en caso contrario. Si utilizamos el mismo truco de retroceso de fase del algoritmo de Deutsch-Jozsa y actuamos sobre un qubit en el estado $|{-}\rangle$, obtenemos la siguiente transformacin:

$$
|x \rangle \xrightarrow{f_s} (-1)^{s\cdot x} |x \rangle
$$

El algoritmo para revelar la cadena de bits oculta se sigue naturalmente al consultar el orculo cuntico $f_s$ con la superposicin cuntica obtenida de la transformacin de Hadamard de $|00\dots 0\rangle$. A saber,

$$
|00\dots 0\rangle \xrightarrow{H^{\otimes n}} \frac{1}{\sqrt{2^n}} \sum_{x\in \{0,1\}^n} |x \rangle \xrightarrow{f_a} \frac{1}{\sqrt{2^n}} \sum_{x\in \{0,1\}^n} (-1)^{a\cdot x}|x \rangle
$$

Debido a que la inversa de las $n$ compuertas Hadamard son nuevamente las $n$ compuertas Hadamard, podemos obtener $a$ mediante

$$
\frac{1}{\sqrt{2^n}} \sum_{x\in \{0,1\}^n} (-1)^{a\cdot x}|x\rangle \xrightarrow{H^ {\otimes n}} |a\rangle
$$

<div class="alert alert-block alert-info">
<strong>Ejemplo</strong>: 
    
Veamos un ejemplo especfico para $n=2$ qubits y una cadena secreta $s=11$. Ten en cuenta que estamos siguiendo la formulacin de la Referencia [J. Du, et.al.](https://arxiv.org/abs/quant-ph/0012114) que genera un circuito para el orculo cuntico de Bernstein-Vazirani utilizando un solo registro.

1. El registro de dos qubits se inicializa a cero:
    
$$\lvert \psi_0 \rangle = \lvert 0 0 \rangle$$

2. Se aplica una compuerta Hadamard a ambos qubits:
    
$$\lvert \psi_1 \rangle = \frac{1}{2} \left( \lvert 0 0 \rangle + \lvert 0 1 \rangle + \lvert 1 0 \rangle + \lvert 1 1 \rangle \right) $$

3. Para la cadena $s=11$, el orculo cuntico realiza la operacin:

$$
|x \rangle \xrightarrow{f_s} (-1)^{x\cdot 11} |x \rangle
$$

$$\lvert \psi_2 \rangle = \frac{1}{2} \left( (-1)^{00\cdot 11}|00\rangle + (-1)^{01\cdot 11}|01\rangle + (-1)^{10\cdot 11}|10\rangle + (-1)^{11\cdot 11}|11\rangle \right)$$

$$\lvert \psi_2 \rangle = \frac{1}{2} \left( \lvert 0 0 \rangle - \lvert 0 1 \rangle - \lvert 1 0 \rangle + \lvert 1 1 \rangle \right) $$

4. Aplicar una compuerta Hadamard a ambos qubits:
    
$$\lvert \psi_3 \rangle = \lvert 1 1 \rangle$$

5. Medir para encontrar la cadena secreta $s=11$

### 2.E.d. Implementacin en Qiskit

Ahora veremos la implementacin del algoritmo Bernstein-Vazirani en Qiskit para una funcin de tres bits con $s=011$.

```python
# importaciones
import matplotlib.pyplot as plt
import numpy as np

from qiskit import IBMQ, Aer
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister, transpile

from qiskit.visualization import plot_histogram
```

Primero establecemos el nmero de qubits utilizados en el experimento y la cadena de bits oculta $s$ que el algoritmo encontrar. La cadena de bits oculta $s$ determina el circuito del orculo cuntico.

```python
n = 3       # nmero de qubits utilizados para representar s
s = '011'   # la cadena binaria oculta
```

Luego usamos Qiskit para programar el algoritmo de Bernstein-Vazirani.

```python
# Necesitamos un circuito con n qubits, ms un qubit auxiliar
# Tambin se necesitan n bits clsicos para escribir la salida
bv_circuit = QuantumCircuit(n+1, n)

# colocarl el qubit auxiliar en el estado |->
bv_circuit.h(n)
bv_circuit.z(n)

# Aplicar compuertas Hadamard antes de consultar al orculo
for i in range(n):
    bv_circuit.h(i)
    
# Aplicar una barrera
bv_circuit.barrier()

# Aplicar el orculo del producto interno
s = s[::-1]        # invertir s para adaptarse al ordenamiento de qubits de qiskit
for q in range(n):
    if s[q] == '0':
        bv_circuit.i(q)
    else:
        bv_circuit.cx(q, n)
        
# Aplicar una barrera
bv_circuit.barrier()

# Aplicar compuertas Hadamard despus de consultar al orculo
for i in range(n):
    bv_circuit.h(i)

# Mediciones
for i in range(n):
    bv_circuit.measure(i, i)

bv_circuit.draw('mpl')
```

#### Simulacin

```python
from qiskit.primitives import Sampler
from qiskit.visualization import plot_histogram

sampler = Sampler()
job_bv = sampler.run(bv_circuit)

legend = ["Resultado BV (valor de s)"]
plot_histogram([job_bv.result().quasi_dists[0].binary_probabilities()], legend=legend)
```

Podemos ver que el resultado de la medicin es la cadena oculta `011`.
