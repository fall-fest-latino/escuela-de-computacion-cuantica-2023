```python
# Ataque tunelamiento

def tuneling_attack(casilla_robo, tableroB,barcosB):
    # Crea un circuito cuántico con 3 qubits
    circ = QuantumCircuit(3)
    # Aplica una puerta X al primer qubit si la casilla de intercambio es un barco
    print("Inputs: ",tableroB[casilla_robo], barcosB)
    if tableroB[casilla_robo] in barcosB:
        print("🟣🔱BARCO TUNELADO 🔱🟣")
        circ.x(0)
        # Aplica una puerta de Hadamard al segundo qubit
        circ.h(1)
        # Aplica puertas CNOT con el segundo y primer qubit como control y el tercer qubit como objetivo
        circ.cx(1, 2)
        circ.cx(0, 1)
        circ.h(0)
        # Ejecuta el circuito en el simulador de estado de vectores
        result = execute(circ, Aer.get_backend('statevector_simulator')).result()
        # Obtiene el estado final del circuito
        state = result.get_statevector(circ)
        return state, tableroB
    else:
        print("🟣 FAIL CUÁNTICO 🟣\n")
        return None, tableroB
```

```python
# Copenhague attack
def copenhague_attack(tablero_oponente):
    num_q = len(tablero_oponente.flatten())
    circ = QuantumCircuit(num_q)
    for i in range(num_q):
        circ.h(i)

    # Aplica la operación de oráculo
    for i in range(num_q):
        if (tablero_oponente.flatten()[i] == 'barco'):
            circ.x(i)
    circ.h(num_q - 1)
    circ.mcx(list(range(num_q - 1)), num_q - 1)  #compuerta multicontrolada x
    circ.h(num_q - 1)
    for i in range(num_q):
        if (tablero_oponente.flatten()[i] == 'barco'):
            circ.x(i)

    # Aplica la operación de difusión
    for qubit in range(num_q):
        circ.h(qubit)
    for qubit in range(num_q):
        circ.x(qubit)
    circ.h(num_q - 1)
    circ.mcx(list(range(num_q - 1)), num_q - 1)  #compuerta multicontrolada x
    circ.h(num_q - 1)
    for qubit in range(num_q):
        circ.x(qubit)
    for qubit in range(num_q):
        circ.h(qubit)

    res = execute(circ, Aer.get_backend('statevector_simulator')).result()
    estado = res.get_statevector(circ)
    estado = np.asarray(estado)

    #calcula las probabilidades de cada estado
    probabilidades = [abs(amplitude)**2 for amplitude in estado]

    #ordena las casillas del tablero por probabilidad
    casillas_ordenadas = sorted(range(num_q), key=lambda i: probabilidades[i], reverse=True)

    #estima las posiciones de los barcos del oponente
    estimacion = set()
    while len(estimacion) < 3:  #asume que hay 4 barcos, el valor se lo puede cambiar por la cantidad de barcos deseada
        estimacion.add(random.choice(casillas_ordenadas[:num_q//2]))

    #convierte los índices a coordenadas
    estimacion_coordenadas = [(i//4, i%4) for i in estimacion]

    #devuelve las posiciones estimadas de los barcos, arroja el número de posiciones como de barcos se estime en el tablero
    return estimacion_coordenadas
```

```python
def bin2dec(bin_):
    decimal = 0
    for digit in bin_:
        decimal = decimal*2 + int(digit)
    return decimal
```

```python
def Qtarget_n(n):
    device_backend = FakeYorktown()
    sim_yorktown = AerSimulator.from_backend(device_backend)
    N_QUBITS = n
    circ = QuantumCircuit(N_QUBITS, N_QUBITS)
    circ.h(0)
    for idx in range(N_QUBITS - 1):
        circ.cx(idx, idx + 1)
    circ.measure_all()
    #circ.draw('mpl')

    tcirc = transpile(circ, sim_yorktown)
    result_noise = sim_yorktown.run(tcirc, shots=1024).result()
    counts_noise = result_noise.get_counts(0)
    #print(counts_noise)
    lcounts=list(counts_noise.keys())
    #print(lcounts)
    for i in range(len(lcounts)):
        lcounts[i]=str(bin2dec(lcounts[i].replace(" 0000",""))+1) # se suma 1 para que los valores vayan desde 0+1 hasta 15+1
    #print(lcounts)
    #return(lcounts[0:4])
    return(lcounts)
```

```python
def Qtarget_():
    circ_ghz = QuantumCircuit(3, 3)     # circuito con 3 qubits y 3 bits clásicos
    circ_ghz.h(0)                       # aplicar compuerta H al qubit 0
    circ_ghz.cx(0, 1)                   # aplicar compuerta CNOT a los qubits 0 y 1
    circ_ghz.cx(0, 2)                   # aplicar compuerta CNOT a los qubits 0 y 2
    circ_ghz.barrier()
    circ_ghz.measure([0,1,2], [0,1,2])  # medir los 3 qubits en los 3 bits clásicos

    #circ_ghz.draw('mpl')                # mostrar el circuito

    simulator = Aer.get_backend('aer_simulator')

    job = execute(circ_ghz, simulator, shots=1)               # ejecutar el circuito una sola vez
    counts = job.result().get_counts(circ_ghz)                # obtener los resultados de la ejecución
    #print(counts)
    r=23
    if '1' in list(counts.keys())[0]:
        r=1
    else:
        r=0
    #print("r=",r)
    return r
```
